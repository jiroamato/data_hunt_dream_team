# Question 1:

```python
import pandas as pd
url = "https://raw.githubusercontent.com/rileywheadon/uss-x-dsci-datathon-datasets/refs/heads/master/bad_weather_1.csv"

df = pd.read_csv(url)
```

```python
df.to_csv('raw_df.csv',index=False)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], format='mixed', errors='raise')
df.loc[df['timestamp'].dt.time == pd.Timestamp('00:00:00').time(), 'timestamp'] += pd.Timedelta(hours=12)
df['timestamp'] = df['timestamp'].dt.strftime('%Y-%m-%d %H:%M')
df
```

```python
df.to_csv('df1.csv',index=False)
```

# Question 2:

```python
url_2 = 'https://raw.githubusercontent.com/rileywheadon/uss-x-dsci-datathon-datasets/refs/heads/master/bad_weather_2.csv'

df2 = pd.read_csv(url_2)
```

```python

df2['timestamp'] = pd.to_datetime(df2['timestamp'], format='mixed', errors='raise')
df2.loc[df2['timestamp'].dt.time == pd.Timestamp('00:00:00').time(), 'timestamp'] += pd.Timedelta(hours=12)
# df2['timestamp'] = pd.to_datetime(df2['timestamp'], format='%Y-%m-%d %H:%M', errors='raise')
# df2['timestamp'] = df2['timestamp'].dt.strftime('%Y-%m-%d %H:%M')
df2
```

```python
from tzfpy import get_tz

df2['timezone'] = df2.apply(
    lambda row: get_tz(row['lon'], row['lat']), 
    axis=1
)
```

```python
def localize_and_convert_to_utc(row):
    
    localized_time = row['timestamp'].tz_localize(row['timezone'], ambiguous='NaT', nonexistent='NaT')
    
    utc_time = localized_time.tz_convert('UTC')
    return utc_time

df2['timestamp_utc'] = df2.apply(localize_and_convert_to_utc, axis=1)
df2['timestamp_utc'] = df2['timestamp_utc'].dt.strftime('%Y-%m-%d %H:%M')
df2['timestamp'] = df2['timestamp'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df2.to_csv('df2.csv',index=False)
```
